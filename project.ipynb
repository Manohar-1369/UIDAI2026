## PATTERN 1: State-Level Enrolment Concentration

```python
# Standardize state names in enrolment data
import matplotlib.pyplot as plt
import numpy as np

df_enrolment["state"] = (
    df_enrolment["state"]
    .astype(str)
    .str.strip()
    .str.lower()
    .replace(state_map)
    .str.title()
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)
```

```python
state_enrolment = df_enrolment.groupby('state')[['age_0_5', 'age_5_17', 'age_18_greater']].sum()
state_enrolment['total'] = state_enrolment.sum(axis=1)
state_enrolment = state_enrolment.sort_values('total', ascending=False)
state_enrolment.head(10)
```

```python
plt.figure(figsize=(12, 6))
top10 = state_enrolment.head(10)
plt.bar(top10.index, top10['total'], color='green')
plt.xticks(rotation=45, ha='right')
plt.title('PLOT 1: Top 10 States by Total Enrolment')
plt.xlabel('State')
plt.ylabel('Total Enrolments')
plt.tight_layout()
plt.show()
```

### 📊 Insight
A small number of states (particularly in Northern and Eastern India) concentrate the majority of enrolment volumes, while peripheral regions show dramatically lower activity.

### 🎯 Why this matters for UIDAI
This geographic concentration creates operational bottlenecks and masks systemic coverage failures in underperforming states.

### ⚙️ Decision or System Improvement Enabled
Audit high-concentration states for data quality and operational efficiency; establish baseline performance targets for bottom 5 states to trigger resource reallocation.

## **ANOMALY DETECTION**
### Finding Unusual Patterns & Red Flags

```python
# Find districts with suspiciously low enrolment
district_enrolment = df_enrolment.groupby(['state', 'district'])[['age_0_5', 'age_5_17', 'age_18_greater']].sum()
district_enrolment['total'] = district_enrolment.sum(axis=1)
district_enrolment = district_enrolment.reset_index()

# Zero enrolment districts
zero_enrolment = district_enrolment[district_enrolment['total'] == 0]
print(f"Districts with ZERO enrolment: {len(zero_enrolment)}")
zero_enrolment.head(10)
```

```python
low_enrolment = district_enrolment.sort_values('total').head(20)
print("Bottom 20 Districts by Enrolment:")
low_enrolment
```

```python
# Calculate day-to-day changes
daily_enrolment['change'] = daily_enrolment['total'].diff()
daily_enrolment['pct_change'] = daily_enrolment['total'].pct_change() * 100

# Find abnormal spikes (>200% increase)
spikes = daily_enrolment[daily_enrolment['pct_change'] > 200]
print(f"Unusual spikes detected: {len(spikes)}")
spikes.head()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(daily_enrolment.index, daily_enrolment['total'], label='Total Enrolment', alpha=0.7)
if len(spikes) > 0:
    plt.scatter(spikes.index, spikes['total'], color='red', s=100, label='Anomalies', zorder=5)
plt.title('Daily Enrolment with Anomalies Highlighted')
plt.xlabel('Date')
plt.ylabel('Enrolments')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 📊 Insight – PLOT 4: Anomaly Detection
Identified multiple enrollment centers with inexplicable spikes (>200% day-to-day growth) and zero-enrollment districts, indicating potential data integrity issues, administrative manipulation, or severe operational failures.

### 🎯 Why this matters for UIDAI
Anomalies undermine data reliability and can mask real enrollment gaps or reveal systemic fraud, both critical for operational oversight.

### ⚙️ Decision or System Improvement Enabled
Implement real-time anomaly alerting: flag spikes >150% for immediate audit; auto-trigger investigation protocol for any district with zero enrollment for 7+ consecutive days.

## **COVERAGE GAP ANALYSIS**
### Comparing Demographic Coverage vs Enrolment Activity

```python
# Total demographic coverage by state
demo_by_state = df_demographic.groupby('state')[['demo_age_5_17', 'demo_age_17_']].sum()
demo_by_state['demo_total'] = demo_by_state.sum(axis=1)

# Merge with enrolment data
coverage_comparison = state_enrolment[['total']].join(demo_by_state['demo_total'], how='outer')
coverage_comparison.columns = ['enrolment', 'demographic']
coverage_comparison = coverage_comparison.fillna(0)
coverage_comparison['gap'] = coverage_comparison['demographic'] - coverage_comparison['enrolment']
coverage_comparison = coverage_comparison.sort_values('gap', ascending=False)
print("Top 10 states with largest coverage gaps:")
coverage_comparison.head(10)
```

```python
plt.figure(figsize=(14, 7))
top_gaps = coverage_comparison.head(15)
x = range(len(top_gaps))
plt.bar(x, top_gaps['demographic'], label='Demographic (Total with Aadhaar)', alpha=0.7)
plt.bar(x, top_gaps['enrolment'], label='Recent Enrolment Activity', alpha=0.7)
plt.xticks(x, top_gaps.index, rotation=45, ha='right')
plt.title('Coverage Gap: Demographic vs Enrolment by State')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

### 📊 Insight – PLOT 5: Coverage Gap Analysis
Stark divergence between demographic population (with Aadhaar) and recent enrollment activity: some states show 1.5M+ population gap, meaning millions with Aadhaar but no active engagement.

### 🎯 Why this matters for UIDAI
Large gaps indicate either: (a) dormant population segments, (b) failed update cycles, or (c) technical barriers to ongoing service access—all critical service failures.

### ⚙️ Decision or System Improvement Enabled
Launch focused update campaigns in high-gap states; establish coverage ratio KPI (>80% of demographic population actively engaging annually); implement outreach to dormant segments.
